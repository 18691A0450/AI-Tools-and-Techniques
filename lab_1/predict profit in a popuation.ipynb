```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd      # to work on the datasets
```

```python
data=pd.read_csv("profitdataset.txt")
print(data.shape)
```

```python
#get x and y data values
x=data['population'].values
y=data['profit'].values
```

```python
# plot the data values
plt.scatter(x,y,40,c='r',label='scatter_data',marker='x')
plt.xlabel('population')
plt.ylabel('profit')
plt.title('food_truck_profit_graph')
plt.grid(True,color='k')
plt.show()
```

```python
# mean values
mean_x=np.mean(x)
mean_y=np.mean(y)
n=len(x)
n
```

```python
# y=mx+c,   to get the values m &c
num=0
den=0
for i in range(n):
    num=num+((x[i]-mean_x)*(y[i]-mean_y))
    den=den+((x[i]-mean_x)**2)
m=num/den
#c=y-m*x
c=mean_y-(m*mean_x)
print(m,c)
```

```python
#caluculate y_predict
y_pred=np.array([])    #to create empty array

for i in range(n):
    p=(m*x[i])+c
    y_pred=np.append(y_pred,np.array([p]),axis=0)
    
print(y_pred)
```

```python
#plot the regression line
plt.plot(x,y_pred,color='m',label='regression_line')
#plot the actual data
plt.scatter(x,y,40,c='g',label='actual data',marker='o')
plt.scatter(x,y_pred,40,c='k',label='predict_data',marker='x')
plt.xlabel('population')
plt.ylabel('profit')
plt.legend()
plt.show()
```

```python
#how good our fit line or regression line is by using r square mean
num=0
den=0
for i in range(n):
    num=num+((y_pred[i]-mean_y)**2) #predicted values of y
    den=den+((y[i]-mean_y)**2)     #actual values of y
r2=num/den
r2   
```
