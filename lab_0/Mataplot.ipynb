# Matplotlib

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

## Linear Graph

```python
x1 = np.array([1,2,3])
y1 = np.array([5,7,4])
x2 = np.array([1,2,3])
y2 = np.array([10,14,12])
plt.plot(x1,y1,label ='first line')
plt.plot(x2,y2,label='second line',color='g')
plt.xlabel('plot number')
plt.ylabel('import var')
plt.title('Intersting graph \n check it out')
plt.legend()
plt.show()
```

# BAR GRAPH

```python
x1 = np.array([2001,2003,2005,2007,2009])
y1 = np.array([500,200,700,800,200])
x2 = np.array([2002,2004,2006,2008,2010])
y2 = np.array([800,600,200,500,1000])
plt.bar(x1,y1,label='dairy milk',color='g')
plt.bar(x2,y2,label='five star',color='m')
plt.xlabel('yearly')
plt.ylabel('production')
plt.title('yearly estimation')
plt.legend()
plt.show()
```

# HISTOGRAM

```python
x =([10,20,30,40,20,40,60,80])
y=([0,20,40,60,80,100,120,140])
plt.hist(x,y,histtype='bar',rwidth=0.8)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Histogram')
plt.legend()
plt.show()
```

# Scatter plot

```python
x =np.array([1,2,3,4,5,6,7])
y =np.array([5,3,4,2,1,5,2])
plt.scatter(x,y,label='star',s=40,marker='*')
plt.xlabel('x',color='r')
plt.ylabel('y',color='r')
plt.legend()
plt.show()
```

# pie chart

```python
s=np.array([7,2,2,13])
plt.pie(s,labels=['sleeping','eating','working','playing'],colors=['c','k','g','b'],startangle=90,shadow='true')
plt.title('timetable')
plt.show()
```

# stack plot/area plot

```python
d=np.array([1,2,3,4,5])
s=np.array([7,8,6,11,7])
e=np.array([2,3,4,3,2])
plt.stackplot(d,s,e,colors=['c','g','k'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Area plot')
plt.show()
```
